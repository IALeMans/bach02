```python
import tarfile
import os
import shutil
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path="../data/"
dataset="idcard3"
```

## recuperation du dataset

```python
#vérifier si le repertoire du dataset est déjà présent
#si oui, suppression
if os.path.isdir(path+dataset):
    shutil.rmtree(path+dataset, ignore_errors=True)
    print(f'repertoire {dataset} effacé')  
else:
    print(f'repertoire {dataset} absent')
```

```python
#verifier la présence du dataset compressé
fname = path+dataset+".tar.gz"
if os.path.exists(path+dataset+".tar.gz"):
    print(f'le fichier {fname} est présent')
else:
    print("ATTENTION, le fichier de données est absent")
```

```python
#extraire le dataset
tar = tarfile.open(fname, "r:gz")
tar.extractall(path=path)
tar.close()
```

```python
#verifier si le repertoire du dataset est bien créé
if os.path.isdir(path+dataset):
    print(f'le repertoire {dataset} est présent. ')  
    print(f'il contient les sous-repertoires suivants : {os.listdir(path+dataset)}')
    
else:
    print(f'ATTENTION, repertoire {dataset} absent')
```

## analyse du dataset

```python
for s in os.listdir(path+dataset):
    print(f'{s} contient {len(os.listdir(path+dataset+"/"+s))} images')
```

## visualisation d'échantillons

#### anything = aucune piece d'identité valide dans l'image

```python
path_det=path+dataset+"/"+os.listdir(path+dataset)[0]
fig = plt.figure(figsize=(20,10))
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    img = plt.imread(path_det+"/"+os.listdir(path_det)[i])
    plt.imshow(img)
```

#### identity = piece d'identité détourée

```python
path_det=path+dataset+"/"+os.listdir(path+dataset)[1]
fig = plt.figure(figsize=(20,10))
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    img = plt.imread(path_det+"/"+os.listdir(path_det)[i])
    plt.imshow(img)
```

#### inside = on présume qu'une pièce d'identité valide est présente dans l'image

```python
path_det=path+dataset+"/"+os.listdir(path+dataset)[2]
fig = plt.figure(figsize=(20,10))
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    img = plt.imread(path_det+"/"+os.listdir(path_det)[i])
    plt.imshow(img)
```
